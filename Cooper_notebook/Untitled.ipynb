```python
tn_budgets['worldwide_gross'] = tn_budgets['worldwide_gross'].str.replace('$','')
```

```python
tn_budgets['worldwide_gross'] = tn_budgets['worldwide_gross'].str.replace(',','')
```

```python
tn_budgets['production_budget'] = tn_budgets['production_budget'].str.replace('$','')
```

```python
tn_budgets['production_budget'] = tn_budgets['production_budget'].str.replace(',','')
```

```python
tn_budgets['worldwide_gross'] = tn_budgets['worldwide_gross'].astype(int)
```

```python
tn_budgets['production_budget'] = tn_budgets['production_budget'].astype(int)
```

```python
tn_budgets = tn_budgets.loc[(tn_budgets['production_budget'] >= 50000) & (tn_budgets['worldwide_gross'] != 0)]
```

We decided to only look at movies that have a production budget of at least 50,000 since we are talking to microsoft and they are most likely going to spend more than 50,000. Also threw out all movies that made $0 since those are either mistakes or extreme outliers.

```python
tn_budgets
```

```python
tn_budgets['release_date'] = pd.to_datetime(tn_budgets['release_date'])
```

```python
tn_budgets
```

```python
tn_budgets['ROI %'] = ((tn_budgets['worldwide_gross']/tn_budgets['production_budget'])*100)
```

```python
tn_budgets = tn_budgets.sort_values('ROI %',ascending=False)
```

```python
tn_budgets
```

```python
tn_budgets = tn_budgets.loc[tn_budgets['ROI %'] > 1]
```

```python
tn_budgets
```

We threw out any movies that had less than 1 percent ROI since in our research, these numbers for worlwide gross were wrong. We also believe that throwing these put still leaves us with enough data.
