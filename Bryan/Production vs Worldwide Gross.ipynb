### Importing the libraries we need.

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
import numpy as np
```

```python
final_table = pd.read_csv('../cleanedData')
```

```python
x = final_table['production_budget']
y = final_table['worldwide_gross']
a, b = np.polyfit(x, y, 1)

plot_six_fig, plot_six_ax = plt.subplots(figsize=(15,6))

plot_six_ax.scatter(x, y)
plot_six_ax.set_title('Production Budget vs Worldwide Gross', fontsize=20)
plot_six_ax.set_xlabel('Production Budget', fontsize=15)
plot_six_ax.set_ylabel('Worldwide Gross', fontsize=15)
plt.plot(x, a*x+b, color='red') 
plot_six_ax.ticklabel_format(useOffset=False, style = 'plain')
```

### As seen above, the relationship between the production budget and worldwide gross is of moderate relationship. As the production budget increase there is going to be an increase in worldwide gross. Which won't always be the case for a couple of outliers.

```python
final_table['profit'] = final_table['worldwide_gross'] - final_table['production_budget']
final_table                                                            
```

```python
x = final_table['production_budget']
y = final_table['profit']
a, b = np.polyfit(x, y, 1)

plot_six_fig, plot_six_ax = plt.subplots(figsize=(15,6))

plot_six_ax.scatter(x, y)
plot_six_ax.set_title('Profit vs Profuction Budget', fontsize=20)
plot_six_ax.set_xlabel('Production Budget', fontsize=15)
plot_six_ax.set_ylabel('Profit', fontsize=15)
plt.plot(x, a*x+b, color='red') 
plot_six_ax.ticklabel_format(useOffset=False, style = 'plain')
```

### As seen above, the relationship between the production budget and profit is of moderate relationship. As the production budget increase there is going to be an increase in proft. Which won't always be the case for a couple of outliers. Regardless, this proves our point that if you have a greater production budget, then you'll have a greater worldwide gross.
